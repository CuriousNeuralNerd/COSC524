### Import Required Libraries

This block imports all the required libraries for data preprocessing, training, and evaluation.

```python
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AutoModelForTokenClassification, pipeline 
import torch
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

#### Entity Removal (NER)

#### Purpose:
- Implement Named Entity Recognition (NER) to anonymize the dataset.
- Replace named entities such as:
  - **Location names**: e.g., cities, countries, or specific geographic references.
  - **Person names**: e.g., character names or author mentions.
  
#### Rationale:
- Prevent the model from relying on specific character or location names, which could bias predictions or reduce generalization performance.
  
#### Implementation:
- Use the `dslim/bert-base-NER` model from Hugging Face Transformers for entity detection.
- Replace detected entities with their respective labels (e.g., `[PER]` for persons, `[LOC]` for locations) in the text.
- Anonymize datasets and provide statistics on the entities detected.


```python
# Load NER model for preprocessing
ner_tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
ner_model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")
ner_pipeline = pipeline("ner", model=ner_model, tokenizer=ner_tokenizer, device=device)

# NER preprocessing function
def extract_entities_and_anonymize(text):
    
    ner_results = ner_pipeline(text)
    extracted_entities = []

    # Replace entities with their labels
    for result in ner_results:
        start, end = result['start'], result['end']
        entity = result['word']
        label = result['entity']
        extracted_entities.append(entity)
        text = text[:start] + f"[{label}] " + text[end:]
    
    return text, extracted_entities

# Apply NER anonymization and extract entities
def process_dataset_with_ner(dataset):

    anonymized_texts = []
    all_entities = []

    for text in dataset['content']:
        anonymized_text, entities = extract_entities_and_anonymize(text)
        anonymized_texts.append(anonymized_text)
        all_entities.extend(entities)

    # Add anonymized content back to the dataset
    dataset['content'] = anonymized_texts

    # Print extracted entities for verification
    print("Extracted Entities:")
    unique_entities = set(all_entities)
    print(unique_entities)
    print(f"Total Entities Extracted: {len(all_entities)}")
    print(f"Unique Entities: {len(unique_entities)}")
    
    return dataset
```

### Load Datasets

Load the training, validation, and test datasets from CSV files.

```python
# Load dataset
train_data = pd.read_csv('./data/train/train_chunked.csv')
valid_data = pd.read_csv('./data/valid/valid_chunked.csv')
test_data = pd.read_csv('./data/test/test_chunked.csv')
```

### Apply NER to Dataset

This block applies NER preprocessing to anonymize named entities in the dataset.

```python
# Apply NER to dataset
train_data = process_dataset_with_ner(train_data)
valid_data = process_dataset_with_ner(valid_data)
test_data = process_dataset_with_ner(test_data)
```

### Combine Data for Tokenization

Split the dataset into features (texts) and labels (categories) for training, validation, and testing.

```python
# Combine data
train_texts, train_labels = train_data['content'], train_data['category']
val_texts, val_labels = valid_data['content'], valid_data['category']
test_texts, test_labels = test_data['content'], test_data['category']
```

### Tokenization

This block tokenizes the text data for BERT, DistilBERT, and XLM-RoBERTa models.

```python
# Tokenization
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
distilbert_tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-multilingual-cased')
xlm_tokenizer = AutoTokenizer.from_pretrained('xlm-roberta-base')

def tokenize_texts(texts, tokenizer, max_length=128):
    return tokenizer(
        list(texts),
        max_length=max_length,
        truncation=True,
        padding='max_length',
        return_tensors='pt'
    )

# Tokenize for each model

# BERT
bert_train_encodings = tokenize_texts(train_texts, bert_tokenizer)
bert_val_encodings = tokenize_texts(val_texts, bert_tokenizer)
bert_test_encodings = tokenize_texts(test_texts, bert_tokenizer)

# DistilBERT
distilbert_train_encodings = tokenize_texts(train_texts, distilbert_tokenizer)
distilbert_val_encodings = tokenize_texts(val_texts, distilbert_tokenizer)
distilbert_test_encodings = tokenize_texts(test_texts, distilbert_tokenizer)

# XLM-RoBERTa
xlm_train_encodings = tokenize_texts(train_texts, xlm_tokenizer)
xlm_val_encodings = tokenize_texts(val_texts, xlm_tokenizer)
xlm_test_encodings = tokenize_texts(test_texts, xlm_tokenizer)
```

### Prepare Dataset for PyTorch

Define a custom dataset class to prepare tokenized data for PyTorch models.

```python
# Prepare Dataset for PyTorch
class CustomDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels.iloc[idx])
        return item

# BERT
bert_train_dataset = CustomDataset(bert_train_encodings, train_labels)
bert_val_dataset = CustomDataset(bert_val_encodings, val_labels)
bert_test_dataset = CustomDataset(bert_test_encodings, test_labels)

# DistilBERT
distilbert_train_dataset = CustomDataset(distilbert_train_encodings, train_labels)
distilbert_val_dataset = CustomDataset(distilbert_val_encodings, val_labels)
distilbert_test_dataset = CustomDataset(distilbert_test_encodings, test_labels)

# XLM-RoBERTa
xlm_train_dataset = CustomDataset(xlm_train_encodings, train_labels)
xlm_val_dataset = CustomDataset(xlm_val_encodings, val_labels)
xlm_test_dataset = CustomDataset(xlm_test_encodings, test_labels)
```

### Define Confusion Matrix Plot

This function visualizes the confusion matrix for model evaluation.

```python
def plot_confusion_matrix(cm, labels, title):
    
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    fig, ax = plt.subplots(figsize=(6, 6))
    disp.plot(ax=ax, cmap='Blues', values_format='d')
    plt.title(title)
    plt.show()
```

### Train and Evaluate Models

This block defines the training process, plots metrics, and evaluates the models using confusion matrices.

```python
def train_and_evaluate(model, train_dataset, val_dataset, test_dataset, model_name):
    # Configure TrainingArguments
    training_args = TrainingArguments(
        output_dir=f'./results_{model_name}',
        evaluation_strategy='epoch',
        save_strategy='epoch',
        logging_strategy='epoch',  # Log only at the end of each epoch
        learning_rate=2e-5,
        per_device_train_batch_size=16,
        per_device_eval_batch_size=32,
        num_train_epochs=4,
        weight_decay=0.01,
        logging_dir=f'./logs_{model_name}',
        load_best_model_at_end=True,
        metric_for_best_model='eval_loss',
        greater_is_better=False,
        report_to="none"  # Disable unnecessary logging
    )

    # Compute metrics
    def compute_metrics(eval_pred):
        logits, labels = eval_pred
        predictions = np.argmax(logits, axis=1)
        precision, recall, f1, _ = precision_recall_fscore_support(labels, predictions, average='binary')
        acc = accuracy_score(labels, predictions)
        return {'eval_accuracy': acc, 'precision': precision, 'recall': recall, 'f1': f1}

    # Initialize Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        compute_metrics=compute_metrics
    )

    # Train the model
    trainer.train()

    # Extract metrics logged at the end of each epoch
    train_log = trainer.state.log_history
    epoch_logs = [log for log in train_log if 'epoch' in log.keys()]

    # Extract losses and accuracies
    train_losses = [log['loss'] for log in epoch_logs if 'loss' in log]
    eval_losses = [log['eval_loss'] for log in epoch_logs if 'eval_loss' in log]
    eval_accuracies = [log['eval_accuracy'] for log in epoch_logs if 'eval_accuracy' in log]

    # Ensure that the number of epochs matches the length of losses and accuracies
    epochs = range(1, len(eval_losses) + 1)

    # Plot metrics
    plt.figure(figsize=(12, 6))

    # Plot loss
    plt.subplot(1, 2, 1)
    plt.plot(epochs, train_losses, label='Training Loss')
    plt.plot(epochs, eval_losses, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f'{model_name} Loss Per Epoch')
    plt.legend()

    # Plot accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, eval_accuracies, label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title(f'{model_name} Validation Accuracy Per Epoch')
    plt.legend()

    plt.tight_layout()
    plt.show()

    # Evaluate the model on the test dataset
    results = trainer.evaluate(test_dataset)
    print(f"\nModel: {model_name}\nTest Evaluation:\n", results)

    # Confusion Matrix
    predictions, labels, _ = trainer.predict(test_dataset)
    pred_labels = np.argmax(predictions, axis=1)
    cm = confusion_matrix(labels, pred_labels)

    # Plot confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["0-Not Agatha", "1-Agatha"])
    disp.plot(cmap='Blues', values_format='d')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.show()
```

## Train and Evaluate BERT

This block trains and evaluates a BERT model using the defined `train_and_evaluate` function.

```python
# Train and Evaluate BERT
bert_model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
train_and_evaluate(bert_model, bert_train_dataset, bert_val_dataset, bert_test_dataset, "BERT")
```

### Train and Evaluate DistilBERT

This block trains and evaluates a DistilBERT model using the defined `train_and_evaluate` function.

```python
# Train and Evaluate DistilBERT
distilbert_model = DistilBertForSequenceClassification.from_pretrained('distilbert-base-multilingual-cased', num_labels=2)
train_and_evaluate(distilbert_model, distilbert_train_dataset, distilbert_val_dataset, distilbert_test_dataset, "DistilBERT")
```

### Train and Evaluate XLM-RoBERTa

This block trains and evaluates an XLM-RoBERTa model using the defined `train_and_evaluate` function.

```python
# Train and Evaluate XLM-RoBERTa
xlm_model = AutoModelForSequenceClassification.from_pretrained('xlm-roberta-base', num_labels=2)
train_and_evaluate(xlm_model, xlm_train_dataset, xlm_val_dataset, xlm_test_dataset, "XLM-RoBERTa")
```
